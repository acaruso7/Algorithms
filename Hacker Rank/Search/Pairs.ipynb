```python
#Use a while loop and two pointer method to remove unessecary checks - passes all test cases
def pairs(k, arr):
    """returns the number of pairs in the array arr whose difference equals the target value k"""
    i = 0
    j = 0
    count = 0
    sorted_arr = sorted(arr)
    while(j < len(arr)):
        diff = sorted_arr[j] - sorted_arr[i]
        if diff < k: #upper bound is too small - increment j
            j += 1
        elif diff > k: #lower bound is too small - increment i
            i += 1
        else:
            j += 1
            count += 1
    return count

print(pairs(2, [1,5,3,4,2]))
```
