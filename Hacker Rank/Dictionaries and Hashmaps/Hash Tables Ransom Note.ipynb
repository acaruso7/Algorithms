```python
def checkMagazine(magazine, note):
    """if every word in note exists in magazine (the proper # of times), return Yes
       if any word in note is missing from magazine, or does not have the proper count, return No"""
    mag_list = magazine.split()
    note_list = note.split()
    word_dict = dict.fromkeys(note_list,0) #dictionary to store the required words from the note, and whether they exist in mag
    for word in note_list:
        word_dict[word] += 1 #count the # of times each word appears in the note
    
    mag_dict = dict.fromkeys(mag_list,0)
    for word in mag_list:
        mag_dict[word] += 1
        
    note_items = list(word_dict.items())
    mag_items = list(mag_dict.items())
    
    for pair in note_items:
        if pair in mag_items:
            pass
        else:
            return 'No'
        
    return 'Yes'


def unit_tests():
    assert checkMagazine('give me one grand today night', 'give one grand today') == 'Yes'
    assert checkMagazine('two times three is not four', 'two times two is four') == 'No'
    assert checkMagazine('ive got a lovely bunch of coconuts', 'ive got some coconuts') == 'No'
    print('test cases passed')
    
unit_tests()
```
