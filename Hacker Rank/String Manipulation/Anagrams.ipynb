```python
def makeAnagram(a,b):
    """returns the minimum # of deletions between two strings a and b to make them anagrams"""
    a_dict = dict.fromkeys(list(a),0)
    for char in list(a):
        a_dict[char] += 1  
    
    b_dict = dict.fromkeys(list(b),0)
    for char in list(b):
        b_dict[char] += 1
    
    seen = [] #list to store the common characters that have already been seen, and deletions accounted for
    count = 0
    for key, value in a_dict.items(): #first check for characters that are only present in one string, and count them
        if key not in b_dict.keys() and key not in seen:
            count += value
        
        elif key in b_dict.keys() and key not in seen:#next check for letters present in both strings,count their diff in frequency
            if value == b_dict[key]: #if the letter occurs with the same frequency in both strings
                pass #no need to delete anthing
            elif value != b_dict[key]: #if the letter occurs with different frequency in the strings
                diff = abs(value - b_dict[key])
                count += diff
            seen.append(key)
    
    #look for letters present in string b, but not present in string a, and count them towards total deletions
    for key, value in b_dict.items():
        if key not in a_dict.keys():
            count += value
    
    return count
    
    
    
    
a = 'cde'
b = 'abc'
# delete d and e from string a
# detele a and b from string b
# 4 total deletions, resulting in two strings consisting of 'c'


print(makeAnagram(a,b))
```
