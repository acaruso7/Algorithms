```python
def maximumToys(prices, k):
    """returns the maximum number of toys that can be bought, given an array prices, and a budget k"""
    sort = sorted(prices)
    spent = 0
    for price in sort:
        if spent < k and spent + price < k:
            spent += price
        else:
            return sort.index(price)

print(maximumToys([1,12,5,111,200,1000,10],50))
```
