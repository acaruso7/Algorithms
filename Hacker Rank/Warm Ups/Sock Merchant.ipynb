```python
def sockMerchant(n, ar):
    unique = set(ar) #unique sock colors
    counts = []
    for color in unique:
        counts.append(ar.count(color)) #count the number of socks of each unique color
        
    num_matches = 0
    for count in counts:
        num_matches += (count // 2) #floor division to find number of matches (pairs of 2) per unique color --> e.g. 3 // 2 = 1
    
    return num_matches
    
    
n = 9
ar = [10, 20, 20, 10, 10, 30, 50, 10, 20]

print(sockMerchant(n, ar))
```
