```python
#BRUTE FORE ATTEMPT - construct the string of length n and count the number of a's . . . this runs into memory overflow

def repeatedString(s, n):
    reps = n // len(s)
    string = s*reps
    remainder = n - len(string)
    remainder = s[:remainder]
    string += remainder
    return string.count('a')
    
    
s = 'aba'
n = 10

print(repeatedString(s, n))
```

```python
#Optimized Solution - instead of consturcting the string of length n, 
#just multiple the count of 'a' in one instance by the number of
#repitions required, and then add any remaining characters

def repeatedString(s, n):
    reps = n // len(s)
    remainder = n - (len(s)*reps)
    extra_chars = s[:remainder] #extra characters if n / s is a decimal
    count = s.count('a')
    count *= reps
    count += extra_chars.count('a')
    return count
    
    
    
s = 'aba'
n = 10

print(repeatedString(s, n))
```
