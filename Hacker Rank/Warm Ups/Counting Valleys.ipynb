```python
def countingValleys(n, s): #a valley is at least 2 consecutive steps below sea level, ending with an up step
    level = 0 #when level = 0, Gary is at sea level
    num_valleys = 0
    for step in s:
        if step == 'U':
            level += 1
            if level == 0: #if taking a step up back to sea level (0), then a valley is just finishing
                num_valleys += 1
        else:
            level -= 1
        
    return num_valleys
    
    
n = 8
s = 'UDDDUDUU'

print(countingValleys(n, s))
```
